```python
import pandas as pd

df = pd.read_csv("merged_final.csv")

# 1. Min and Max acq_date
df['acq_date'] = pd.to_datetime(df['acq_date'], errors='coerce')
print("Date range in acq_date:")
print("  Minimum date:", df['acq_date'].min())
print("  Maximum date:", df['acq_date'].max())
print()

# 2. Check for duplicates of (lat, lon, acq_date)
dup_count = df.duplicated(subset=['latitude','longitude','acq_date']).sum()
print(f"Number of duplicate (latitude, longitude, acq_date) rows: {dup_count}")
print()

# 3. Unique values in 'version'
if 'version' in df.columns:
    print("Unique values in 'version':")
    print(df['version'].value_counts())
    print()

# 4. Correlation matrix
#    We'll limit columns to numeric ones for clarity.
num_cols = ['latitude','longitude','brightness','frp','lat_round','lon_round','NDVI']
corr_matrix = df[num_cols].corr()
print("Correlation Matrix (numeric columns):")
print(corr_matrix)
```

```python
import os
import rasterio
import numpy as np

# Change this to the path of ONE of your .tif files
tif_file = "./tif/MOD13A2.061__1_km_16_days_blue_reflectance_doy2019209_aid0001.tif"

with rasterio.open(tif_file) as src:
    ndvi_data = src.read(1)
    ndvi_data = np.ma.masked_equal(ndvi_data, src.nodata)

    print("Raw NDVI statistics from one TIF:")
    print("  Min:", ndvi_data.min())
    print("  Max:", ndvi_data.max())
    print("  Mean:", ndvi_data.mean())

# You should see something like:
# Min: -3000
# Max: 10000
# or
# Max: 63487
# etc.
```

```python
import pandas as pd

df = pd.read_csv("merged_final.csv")

df['NDVI'] = df['NDVI'] / 10000.0

df.to_csv("merged_final_rescaled.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_final_rescaled.csv")  # the new CSV after dividing by 10000

print("Descriptive Statistics (after NDVI rescaling):")
print(df[['brightness','frp','NDVI']].describe())

print("\nCorrelation Matrix (after NDVI rescaling):")
print(df[['brightness','frp','NDVI']].corr())

# Quick histogram for NDVI after rescaling
plt.figure(figsize=(6,4))
df['NDVI'].hist(bins=50, edgecolor='black')
plt.title('Rescaled NDVI Distribution')
plt.xlabel('NDVI')
plt.ylabel('Frequency')
plt.show()
```
